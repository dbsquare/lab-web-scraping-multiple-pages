## Lab | Web Scraping Single Page

- Songs that are actually similar to the ones they picked from an acoustic point of view.
- Songs that are popular around the world right now, independently from their tastes.

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
pd.set_option('display.max_rows', None)
```

```python
r = requests.get('http://www.popvortex.com/music/charts/top-100-songs.php')
r.status_code
```

```python
html = r.content
```

```python
p = BeautifulSoup(r.text, 'html.parser')
```

```python
titles= []

for i in p.find_all(class_='title'):
    titles.append(i.get_text())
titles
```

```python
artists= []

for i in p.find_all(class_='artist'):
    artists.append(i.get_text())
artists
```

```python
df = pd.DataFrame({'artist':artists,'titles':titles})
df
```

```python
len(bb.select('h3.c-title.a-no-trucate'))
from tqdm import tqdm
```

```python
# we create empty lists where we will store info later below 
artists_billboard = []
songs_billboard = []

#check how many times the loop has to run
list_billboard = len(bb.select('h3.c-title.a-no-trucate'))


# the loop to populate the lists
for i in tqdm(range(list_billboard)):
    songs_billboard.append(bb.select('h3.c-title.a-no-trucate')[i].get_text(strip = True))
    artists_billboard.append(bb.select('span.c-label.a-no-trucate.a-font-primary-s')[i].get_text(strip = True))
```

```python
df = pd.DataFrame({'song':songs_billboard, 'author':artists_billboard})
df
```
